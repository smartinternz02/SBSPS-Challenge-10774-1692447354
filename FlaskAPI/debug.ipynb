```python
# Function to convert an existing image file to FileStorage object
def convert_image_to_filestorage(image_path):
    with open(image_path, 'rb') as file:
        file_storage = FileStorage(
            stream=file,
            filename=file.name,
            content_type='image/jpeg'  # Update with the appropriate content type
        )
    print(file_storage)
    return file_storage
```

```python
def store_img_in_dir(temp_dir_path, file):
    # Ensure the directory exists before saving the file
    os.makedirs(temp_dir_path, exist_ok=True)

    # Secure the filename to prevent any potential security issues
    filename = secure_filename(file.filename)

    # Construct the destination path
    destination_path = os.path.join(temp_dir_path, filename.split('\\')[-1])

    # Print for debugging
    print(f"Saving {file.filename} to {destination_path}")

    # Reopen the file and save it to the destination path
    with open(destination_path, 'wb') as dest_file:
        dest_file.write(file.read())
```
